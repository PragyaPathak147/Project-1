```python
X=df[['pop','poplt5','pop5_17','pop18p','pop65p','popurban','medage','marriage','divorce']]
X.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
Scaler= StandardScaler()
X_scaled= Scaler.fit_transform(X)
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
X_scaled.head()
```

```python
import numpy as np
X_scaled['istrain']=np.random.uniform(0,1,len(df))<=0.75
X_scaled.head()
X_scaled['death']=df['death']
```

```python
X_scaled['death']=df['death']
X_scaled.head()
```

```python
X_scaled_train=X_scaled[X_scaled['istrain']==True]
X_scaled_test=X_scaled[X_scaled['istrain']==False]
```

```python
Y_train=X_scaled_train['death']
Y_test=X_scaled_test['death']
```

```python
X_scaled_train=X_scaled_train.drop(['death','istrain'],axis=1)
X_scaled_test=X_scaled_test.drop(['death','istrain'],axis=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_scaled_train,Y_train)
Y_pred=lr.predict(X_scaled_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
